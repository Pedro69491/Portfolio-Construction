# Asset Frontier

```python
n_points = 20
weights = [np.array((w,1-w)) for w in np.linspace(0,1, n_points)]
```

```python
weights
```

```python
vol = [tk.portfolio_volatility(w, cov.loc[['Games', 'Fin'], ['Games', 'Fin']]) for w in weights]
ret = [tk.portfolio_return(w, tk.annualized_return(rets[['Games', 'Fin']],12)) for w in weights]
```

```python
ef = pd.DataFrame({'Return':ret, 'Volatility':vol})
ef.plot.scatter(x='Volatility', y='Return')
```

```python
ef.plot.line(x='Volatility', y='Return', style=".-")
```

# Multi-Asset efficient-frontier

```python
tk.plot_ef(cov.loc[l,l], er, 25)
```

# Capital Market Line

```python
er = tk.annualized_return(rets['1996':'2000'], 12)
```

```python
ax = tk.plot_ef(cov, er, 20)
ax.set_xlim(left=0)
rfr = 0.1
w_msr = tk.msr(rfr, cov, er)
vol_msr = tk.portfolio_volatility(w_msr, cov)
r_msr = tk.portfolio_return(w_msr, er)
x_vals = [0, vol_msr]
y_vals =[rfr, r_msr]

ax.plot(x_vals, y_vals, color='green', linestyle='--', marker='o')
```
